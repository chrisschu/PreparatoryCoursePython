With print(), we can display different outputs.

We can also display the values of variables




```python
# In the example you can see that the values ​​25 and "Gustav" are different

# age is a number
# name is a text (recognizable by " " or ' ')

age = 25
name = "Gustavo"

print(age)
print(name)
```

### String Concatenation



```python
firstName = "Gustav"
lastName = "Ganz"

#1 option with the sign ,
fullName = firstName,   lastName
print(fullName)
```

```python
# 2 Möglichkeit mit +
fullName = firstName + " " + lastName
print(fullName)
```

## How can numbers and letters be combined in Python?

```python
# Simple print output once directly the value 43
x = 42 # We set x to 43
print("The value of the variable x is: ", x)
```

```python
# Output without line breaks
# We want to print the value of the variable x without a line break after it.
x = 43 # We set the value of x to 42
print("The value of the variable x is: ", end="") # end="" prevents the line break
print(x)
```

```python
# Output with line break
# Here we output a text that has a line break at the end. This happens automatically.
print("\nPrint a line break at the end") # \n inserts a new line
```

```python
# Output with formatting (using .format())
# Using the format() method, we can easily insert variables into a string.
print("The value of the variable x is: {}".format(x)) # {} is a placeholder for x
```

```python
# Alternatively, we can use so-called "f-strings" (since Python 3.6).
# This is often the easiest way to insert variables into strings.
print(f"The value of variable x is: {x}") # f in front of the string allows us to insert variables directly
```
